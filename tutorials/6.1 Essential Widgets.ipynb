## `interact` for Instant Widgets

`interact` is a brilliant piece of functionality that makes widgets accessible to anyone with basic Python skills. It does require an import `from ipywidgets import interact`

When you call `interact` just pass in a function, and the names of the arguments to that function you would like to interact with, along with a range of values, and `interact` will automatically generate a widget that allows you to change the input values and dynamically see the output value.

### A simple example

```python
# take a value x and return x squared
def square(x):
    return x*x
```

```python
interact(square, x=(0,100))
```

### A teaching example

```python
#generate and plot a sine wave with frequency freq, and amplitude amp
def make_sine(freq, amp):
    plt.xlim(0, 1)
    plt.ylim(-10, 10)
    t = np.linspace(0, 1, 1000)
    plt.plot(t, amp*np.sin(freq*2*np.pi*t))
```

```python
make_sine(10, 5)
```

Of course we could make 3 or 4 of these example calls to make_sine with parameters chosen to demonstrate the effects of altering frequency and amplitude. This is the approach taken by every textbook you've ever read. Unfortunately, textbooks are static. They'd be interactive if they could, but they can't, but Jupyter can.

```python
interact(make_sine, freq=(1,20), amp=(1,10))
```

Instead of the textbook author deciding on the correct number of examples and parameters for the student to understand the effects of altering frequency and amplitude, the user gets to easily experience any combination they feel they need to internalize the concept at hand.

### Custom Ranges

```python
def hear_sine(freq, duration):
    t = np.linspace(0, duration, duration*16000)
    display(Audio(np.sin(freq*2*np.pi*t), rate=16000));
```

```python
hear_sine(100, 2)
```

```python
interact(hear_sine, freq=(100,1000,50), duration=(1,10,1))
```
